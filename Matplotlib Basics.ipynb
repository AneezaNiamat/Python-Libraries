```
import matplotlib.pyplot as plt
import pandas as pd
```

```
days=[1,2,3,5,6,7]# NO. OF WEEK DAYS
temp=[50,51,48,47,42,40]#temperature
```

```
plt.plot(days,temp,color='red',linestyle='dotted',marker='o',markerfacecolor='purple')
plt.ylabel('Temp')
plt.xlabel('No. of weeks')
plt.title('Weather Report')
plt.show()
```

```
plt.scatter(days,temp,color='red',marker='*')#For scatter graph
plt.ylabel('Temp')
plt.xlabel('No. of weeks')
plt.title('Weather Report')
plt.show()
```

```
plt.bar(days,temp,color='red')
plt.ylabel('Temp')
plt.xlabel('No. of weeks')
plt.title('Weather Report')
plt.show()
```

```
mylabels = ["Monday", "Tuesday", "Wednesday", "Thursday",'Friday',"Saturday"]
plt.pie(days, labels = mylabels)#pass only x-axis labels
# plt.ylabel('Temp')
# plt.xlabel('No. of weeks')
plt.title('Weather Report')
plt.show()
```
